```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('funcionarios_dados_ti (1).csv', sep= ';')
df
```

```python
#separar só o tempo de espera
valor= df['Salario']
dados= np.array(valor)
dados
```

```python
#calcular média
media= np.mean(dados)
media
```

```python
#calcular mediana
mediana= np.median(dados)
mediana
```

```python
#calcular a distancia relativa
distancia= (media - mediana) / mediana
distancia
```
