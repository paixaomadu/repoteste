```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv', sep= ';', encoding='latin1')
df
```

```python
df_cap= df.loc[(df['regiao'] == 'Capital') & (df['ano'] >= 2015)]
df_cap
```

```python
df_cap= df_cap.groupby(['ano', 'cisp', 'munic', 'regiao'])[['roubo_veiculo', 'furto_veiculos', 'registro_ocorrencias']].sum().reset_index()
df_cap= df_cap.sort_values(by= 'ano')
df_cap
```
